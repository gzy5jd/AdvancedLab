```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = !ls *.txt
print(x)

labels = ['Ambient', 'Ar', 'CO2', 'He', 'N2', 'O2']
```

```python
labels[0]
```

```python
def plot(n):
    dat = np.loadtxt(x[n])
    plt.plot(dat[:,0], (dat[:,1] - np.min(dat[:,1])) / np.max(dat[:,1]),label=labels[n])
    plt.scatter(dat[:,0], (dat[:,1] - np.min(dat[:,1])) / np.max(dat[:,1]),s=10)
    plt.legend()


for i in range(len(x)):
    plot(i)

plt.xlim(1775,2100)
plt.ylabel('Intensity (a.u.)')
plt.grid()
```
